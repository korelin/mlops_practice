## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Data generation

```python
SEED = 42
```

```python
np.random.seed(seed=SEED)
xs = np.linspace(0, 10, 100)
```

```python
ys1 = xs + np.random.random(100)*2 - 1

plt.scatter(xs, ys1)
plt.title("Dataset 1")
plt.show()
```

```python
ys2 = xs + np.random.random(100)*2 - 1

plt.scatter(xs, ys2)
plt.title("Dataset 2")
plt.show()
```

```python
ys3 = xs + np.random.random(100)*2 - 1

plt.scatter(xs, ys3)
plt.title("Dataset 3")
plt.show()
```

```python
ys4 = xs + np.random.random(100)*2 - 1
ys4[25:45] *= 2

plt.scatter(xs, ys4)
plt.title("Dataset 4")
plt.show()
```

## Learn Model

```python
reg = LinearRegression()
reg.fit(xs.reshape(-1, 1), ys1)
```

```python
## Predict for first dataset
```

```python
predicted_y = reg.predict(xs.reshape(-1, 1))
```

## Save datasets

```python
np.savetxt("xs.csv", xs, delimiter=",")

np.savetxt("ys1.csv", ys1, delimiter=",")
np.savetxt("ys2.csv", ys2, delimiter=",")
np.savetxt("ys3.csv", ys3, delimiter=",")
np.savetxt("ys4.csv", ys4, delimiter=",")

np.savetxt("predicted_y.csv", predicted_y, delimiter=",")
```

# Draw datasets

```python
plt.scatter(xs, ys1)
plt.plot(xs.reshape(-1, 1), predicted_y, color="blue", linewidth=3)
plt.show()
print("Mean squared error: %.4f" % mean_squared_error(ys1, predicted_y))
print("Coefficient of determination: %.4f" % r2_score(ys1, predicted_y))
print("Coefficients: ", reg.coef_)
```

```python
plt.scatter(xs, ys2)
plt.plot(xs.reshape(-1, 1), predicted_y, color="blue", linewidth=3)
plt.show()
print("Mean squared error: %.4f" % mean_squared_error(ys2, predicted_y))
print("Coefficient of determination: %.4f" % r2_score(ys2, predicted_y))
print("Coefficients: ", reg.coef_)
```

```python
plt.scatter(xs, ys3)
plt.plot(xs.reshape(-1, 1), predicted_y, color="blue", linewidth=3)
plt.show()
print("Mean squared error: %.4f" % mean_squared_error(ys3, predicted_y))
print("Coefficient of determination: %.4f" % r2_score(ys3, predicted_y))
print("Coefficients: ", reg.coef_)
```

```python
plt.scatter(xs, ys4)
plt.plot(xs.reshape(-1, 1), predicted_y, color="blue", linewidth=3)
plt.show()
print("Mean squared error: %.4f" % mean_squared_error(ys4, predicted_y))
print("Coefficient of determination: %.4f" % r2_score(ys4, predicted_y))
print("Coefficients: ", reg.coef_)
```
